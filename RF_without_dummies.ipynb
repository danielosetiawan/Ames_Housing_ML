```python
 RF2 = RandomForestRegressor(n_jobs=-1, max_depth = 10, min_samples_leaf = 2, n_estimators = 500, min_samples_split = 5, random_state=0)
```

```python
RF2.fit(X_train, y_train)
print(f'The R-squared on test set is {RF2.score(X_test,y_test)}')
print(f'The R-squared on train set is {RF2.score(X_train,y_train)}')
```

```python
## Fixing the hyperparameters (grid search is very expensive), we'll do some feature reduction by looking at permuation importance.

from sklearn.inspection import permutation_importance


ls = []
feat_list = []
Xtest_list = [X_test]
Xtrain_list = [X_train]

```

```python
pd.set_option('display.max_columns', 100)
param_grid= {'max_depth': [10],
                         'min_samples_leaf': [2],
                         'min_samples_split': [5],
                         'n_estimators': [500]}

grid = GridSearchCV(RandomForestRegressor(n_jobs=-1, 
                                    random_state=0), 
                    param_grid = param_grid, 
                    cv=5).fit(Xtrain_list[56], y_train)X_test
```

```python
import warnings
warnings.filterwarnings('ignore')

for j in range(0,85):
    results = permutation_importance(RF2.fit(Xtrain_list[j], y_train), 
                                     Xtest_list[j], 
                                     y_test, 
                                     scoring='neg_mean_squared_error')
    # get importance
    sorted_idx = results.importances_mean.argsort()

    feature_rank = {}
    features = np.array(Xtest_list[j].columns.values)
    feat_list.append(features)
    # summarize feature importance
    for i in sorted_idx[::-1]:
        data = {features[i]: results.importances_mean[i]}
        feature_rank.update(data)
    
    last_feat = list(feature_rank)[-1]
    del feature_rank[last_feat]
    ls.append(RandomForestRegressor(n_jobs=-1, 
                                    max_depth = 10, 
                                    min_samples_leaf = 2, 
                                    n_estimators = 500, 
                                    min_samples_split = 5,
                                    random_state=0)
              .fit(Xtrain_list[j], y_train))
    Xtest_list.append(X_test[list(feature_rank)])
    Xtrain_list.append(X_train[list(feature_rank)]) 
```

```python
df1 = pd.DataFrame(columns=['R-sq', 'Num_of_feat'])
for i in range(0,85):
    df1=df1.append({'R-sq':ls[i].score(Xtest_list[i], y_test), 'Num_of_feat': 85-i}, ignore_index=True)
```

```python
from matplotlib import pyplot as plt
fig = plt.figure()
plt.plot(df1['Num_of_feat'], df1['R-sq'])
plt.xlabel('Number of model features')
plt.ylabel('R-square value')
fig.savefig('rf_feat_imp.png', dpi=1600)
```
